# **1. Defanging an IP Address :**

```python
def defang_ip_address(address: str) -> str:
    """
    Replace every '.' with '[.]' in the given IP address string.
    Example: "1.1.1.1" -> "1[.]1[.]1[.]1"
    """
    return address.replace('.', '[.]')

# Example
print(defang_ip_address("1.1.1.1"))
```

# **2. Find Numbers with an Even Number of Digits :**

```python
def find_numbers_with_even_digits(nums):
    """
    Return count of numbers in nums that have even number of digits.
    Example: [12,345,2,6,7896] -> 2 (12 and 7896)
    """
    count = 0
    for x in nums:
        # convert to string to count digits (works for non-negative ints)
        if len(str(abs(x))) % 2 == 0:
            count += 1
    return count

# Example
print(find_numbers_with_even_digits([12, 345, 2, 6, 7896]))
```

# **3. Number of Good Pairs :**

```python
def num_good_pairs_bruteforce(nums):
    """
    Count pairs (i, j) with i < j and nums[i] == nums[j]
    using a brute-force approach.
    """
    count = 0
    n = len(nums)
    for i in range(n):
        for j in range(i + 1, n): # Start j from i+1 to ensure i < j
            if nums[i] == nums[j]:
                count += 1
    return count

# Example
print(num_good_pairs_bruteforce([1,2,3,1,1,3]))
```

# **4. How Many Numbers Are Smaller Than the Current Number :**

```python
def smaller_numbers_than_current(nums):
    """
    For each element find how many numbers are smaller than it.
    Efficient: sort unique values, map value -> count of numbers smaller.
    """
    # sorted list of nums
    sorted_nums = sorted(nums)
    # map value -> first index in sorted order (count of smaller)
    first_index = {}
    for i, val in enumerate(sorted_nums):
        if val not in first_index:
            first_index[val] = i
    # map each original number to its count
    return [first_index[x] for x in nums]

# Example
print(smaller_numbers_than_current([8,1,2,2,3]))
```

# **5. Subtract the Product and Sum of Digits of an Integer :**

```python
def subtract_product_and_sum(n: int) -> int:
    """
    Return product of digits minus sum of digits.
    Example: n=234 -> (2*3*4) - (2+3+4) = 24 - 9 = 15
    """
    prod = 1
    summ = 0
    x = abs(n)
    if x == 0:
        return 0 - 0  # product 0, sum 0
    while x > 0:
        d = x % 10
        prod *= d
        summ += d
        x //= 10
    return prod - summ

# Example
print(subtract_product_and_sum(4421))
```

# **6. XOR Operation in an Array :**

```python
def xor_operation(n: int, start: int) -> int:
    """
    Construct array arr[i] = start + 2*i for i in [0..n-1], return XOR of all elements.
    Example: n=5, start=0 -> arr=[0,2,4,6,8] -> xor = 8
    """
    result = 0
    for i in range(n):
        result ^= (start + 2 * i)
    return result

# Example
print(xor_operation(5, 0))  # Output: 8
print(xor_operation(4, 3))  # arr=[3,5,7,9] -> xor= (3^5^7^9) -> prints result
```
