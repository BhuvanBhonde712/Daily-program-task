# **1. Maximum Product of Two Elements in an Array :**

```python
def max_product(nums):
    nums.sort()
    return (nums[-1] - 1) * (nums[-2] - 1)

# Example
print(max_product([3, 4, 5, 2]))
```

# **2. Count Number of Teams :**

```python
def num_teams(rating):
    n = len(rating)
    count = 0

    for j in range(n):
        left_smaller = left_greater = 0
        right_smaller = right_greater = 0

        # left side
        for i in range(j):
            if rating[i] < rating[j]:
                left_smaller += 1
            elif rating[i] > rating[j]:
                left_greater += 1

        # right side
        for k in range(j + 1, n):
            if rating[k] < rating[j]:
                right_smaller += 1
            elif rating[k] > rating[j]:
                right_greater += 1

        # increasing + decreasing combinations
        count += left_smaller * right_greater
        count += left_greater * right_smaller

    return count

# Example
print(num_teams([2,5,3,4,1]))
```

# **3. Number of Students Doing Homework at a Given Time :**

```python
def busy_students(startTime, endTime, queryTime):
    count = 0
    for s, e in zip(startTime, endTime):
        if s <= queryTime <= e:
            count += 1
    return count

# Example
print(busy_students([1,2,3], [3,2,7], 4))
```

# **4. Number of Steps to Reduce a Number to Zero :**

```python
def number_of_steps(num):
    steps = 0
    while num > 0:
        if num % 2 == 0:
            num //= 2
        else:
            num -= 1
        steps += 1
    return steps

# Example
print(number_of_steps(14))
```

# **5. Counting Bits :**

```python
def count_bits(n):
    dp = [0] * (n + 1)

    for i in range(1, n + 1):
        dp[i] = dp[i >> 1] + (i & 1)

    return dp

# Example
print(count_bits(5))
```
